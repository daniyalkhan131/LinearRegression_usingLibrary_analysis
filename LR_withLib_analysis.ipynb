```python
import random as r
import matplotlib.pyplot as plt
ds=list()
for i in range(100):
    temp=list()
    temp.append(r.randrange(1,50))
    temp.append(r.randrange(30,70))
    ds.append(temp)
print(ds)
xc=list()
yc=list()
for i in ds:
    xc.append(i[0])
    yc.append(i[1])
plt.plot(xc,yc,"bo")

```

```python
j=0
i=0
parts=list()
while(i<100):
    i=i+10
    parts.append(ds[j:i])
    j=i
print(parts)
print('\n')

test_set=list()
for i in parts:
    test_set.append(i[8])
    test_set.append(i[9])
    i.remove(i[8])
    i.remove(i[8])
print(test_set)
train_set=list()
train_set=parts
print('\n')
print(train_set)
```

```python
from sklearn.linear_model import LinearRegression

X_test=list()
y_test=list()
for i in test_set:
    X_test.append([i[0]])
    y_test.append([i[1]])
    
predicted_value=list()
    
for pas in range(len(train_set)):
    X=list()
    y=list()
    for k in range(0,pas+1):
        for i in range(len(train_set[pas])):
            X.append([train_set[k][i][0]])
            y.append([train_set[k][i][1]])
    reg = LinearRegression()
    reg.fit(X,y)
    print('\n')
    print('m'+str(pas))
    print(X,y)
    predicted_value.append(reg.predict(X_test))
```

```python
predicted_value
```

```python
def meanSquare(l1,l2):
    sum=0
    for i,j in  zip(l1,l2):
        sum=sum+(i[0]-j[0])**2
    sum=(sum)/len(l1)
    return sum
```

```python
error_list=list()
for i in predicted_value:
    error_list.append(meanSquare(i,y_test))
error_list
```

```python
import matplotlib.pyplot as plt
z=list()
for i in range(10):
    z.append(i+1)
plt.plot(z,error_list,'.r')
plt.plot(z,error_list,'-b')
```
